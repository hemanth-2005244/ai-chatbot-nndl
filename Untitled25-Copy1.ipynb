```python
import tensorflow as tf
import numpy as np
import re
import pickle
```

```python
data = [
    ("hi", "hello"),
    ("hello", "hi"),
    ("how are you", "i am fine"),
    ("what is your name", "i am a chatbot"),
    ("who are you", "i am an ai assistant"),
    ("what do you do", "i help people"),
    ("bye", "goodbye"),
    ("thanks", "welcome"),
]
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

inputs = []
outputs = []

for inp, out in data:
     inputs.append(clean_text(inp))
     outputs.append(clean_text(out))
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(inputs + outputs)

input_seq = tokenizer.texts_to_sequences(inputs)
output_seq = tokenizer.texts_to_sequences(outputs)

input_seq = tf.keras.preprocessing.sequence.pad_sequences(input_seq, padding='post')
output_seq = tf.keras.preprocessing.sequence.pad_sequences(output_seq, padding='post')

vocab_size = len(tokenizer.word_index) + 1
max_len = input_seq.shape[1]
```

```python
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
from tensorflow.keras.models import Model

inputs_layer = Input(shape=(max_len,))
x = Embedding(vocab_size, 64, mask_zero=True)(inputs_layer)
x = LSTM(128)(x)
outputs_layer = Dense(vocab_size, activation='softmax')(x)

model = Model(inputs_layer, outputs_layer)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

```python
X = input_seq
y = np.array([seq[0] for seq in output_seq])
```

```python
model.fit(X, y, epochs=300, verbose=0)
```

```python
model.save("chatbot_model.keras")

with open("tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)
```

```python
def chatbot_reply(text):
    text = clean_text(text)
    seq = tokenizer.texts_to_sequences([text])
    seq = tf.keras.preprocessing.sequence.pad_sequences(seq, maxlen=max_len)

    pred = model.predict(seq, verbose=0)
    word_id = np.argmax(pred)

    return tokenizer.index_word.get(word_id, "I don't understand")
```

```python
print(chatbot_reply("hi"))
print(chatbot_reply("bye"))
```
